#### 🔹 1. Sending a Basic HTTP Request

Write a Python script using requests to fetch the HTML content of a webpage and print the status code.

````python
import requests

url = "https://example.com"
response = requests.get(url)

print("Status Code:", response.status_code)
print(response.text[:500])
````

#### 🔹 2. Parsing HTML with **BeautifulSoup**

Extract the title of a webpage using BeautifulSoup.

````python
from bs4 import BeautifulSoup
import requests

url = "https://example.com"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
title = soup.title.text

print(title)
````

#### 🔹 3. Extract All Links from a Page

Write a script to extract all `<a>` tag links.

````python
links = soup.find_all("a")

for link in links:
    print(link.get("href"))
````

#### 🔹 4. Extract Data Using Class Name

Extract all product names using class `"product-title"`

````python
products = soup.find_all("div", class_="product-title")

for p in products:
    print(p.text.strip())
````

#### 🔹 5. Using CSS Selectors

Extract elements using `.select()`

````python
titles = soup.select(".product-title")

for t in titles:
    print(t.text)
````

#### 🔹 6. Nested HTML Extraction

Extract product name and price from this structure:

```python
<div class="product">
    <h2>Item A</h2>
    <span class="price">$10</span>
</div>
```

````python
# <div class="product">
#     <h2>Item A</h2>
#     <span class="price">$10</span>
# </div>
````

````python
products = soup.find_all("div", class_="product")

for p in products:
    name = p.find("h2").text
    price = p.find("span", class_="price").text
    print(name, price)
````

#### 🔹 7. Handling Missing Elements (IMPORTANT)

Sometimes elements don’t exist. Handle safely.

````python
price = p.find("span", class_="price")

if price:
    print(price.text)
else:
    print("Price not available")
````

#### 🔹 9. Pagination Scraping (VERY IMPORTANT)

Scrape multiple pages.

````python
for page in range(1, 6):
    url = f"https://example.com/page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    titles = soup.select(".product-title")

    for t in titles:
        print(t.text)
````

#### 🔹 10. Rate Limiting (IMPORTANT)

Avoid getting blocked.

👉 **Interview Tip:**
Say: “I always add delay to mimic human behavior”

````python
import time

for i in range(5):
    print("Scraping page", i)
    time.sleep(2)
````

#### 🔹 13. Selenium: Wait for Elements (VERY IMPORTANT)

Use explicit wait.

````python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)

element = wait.until(
    EC.presence_of_element_located((By.CLASS_NAME, "product"))
)
````

#### 🔹 15. Handle "Load More" Button

#### 🔹 20. Real Interview Question

Write a scraper that:

1. Scrapes titles from a website
2. Handles pagination
3. Uses headers
4. Adds delay

````python
import requests
from bs4 import BeautifulSoup
import time

headers = {"User-Agent": "Mozilla/5.0"}

for page in range(1, 4):
    url = f"https://example.com/page={page}"
    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")
    titles = soup.select(".product-title")

    for t in titles:
        print(t.text.strip())

    time.sleep(2)
````
