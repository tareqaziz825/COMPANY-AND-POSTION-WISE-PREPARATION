### E. APIs & FastAPI (IMPORTANT)

#### 🔹 1. What is an API?

API (Application Programming Interface) is a way for two applications to communicate with each other. It allows one system to request data or services from another system.

#### 🔹 2. What is a REST API?

A REST API follows REST (Representational State Transfer) principles and uses HTTP methods like GET, POST, PUT, DELETE to perform operations.

#### 🔹 3. What is the difference between GET and POST?

- **GET:** Used to retrieve data from a server.
- **POST:** Used to send data to a server to create/update resources.

#### 🔹 4. What is JSON?

JSON (JavaScript Object Notation) is a lightweight data format used for data exchange between client and server. It is easy to read and write.

#### 🔹 5. Why is JSON commonly used in APIs?

Because it is lightweight, human-readable, and easily parsed by most programming languages including Python.

#### 🔹 6. How do you make an API request in Python?

Using the `requests` library:

```python
import requests

response = requests.get("https://api.example.com")
data = response.json()
```

#### 🔹 7. What is FastAPI?

FastAPI is a modern Python web framework used to build APIs quickly and efficiently with high performance.

#### 🔹 8. Basic FastAPI Example

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello"}
```

#### 🔹 9. How do you run a FastAPI application?

Using Uvicorn:

```bash
uvicorn main:app --reload
```

#### 🔹 10. What is path operation in FastAPI?

A path operation is a function that is executed when a specific API endpoint is accessed (e.g., `@app.get("/")`).

#### 🔹 11. What is request and response in API?

- **Request:** Sent by client to server
- **Response:** Sent by server back to client (usually JSON)

#### 🔹 12. How can APIs be useful in scraping/automation?

APIs provide structured data directly, making scraping easier, faster, and more reliable than parsing HTML.

### Error Handling & Debugging (VERY IMPORTANT)


#### 🔹 13. What is exception handling in Python?

It is a way to handle runtime errors using `try-except` blocks without crashing the program.

#### 🔹 14. Basic try-except example

```python
try:
    x = 10 / 0
except Exception as e:
    print(e)
```

#### 🔹 15. Why is error handling important in scraping?

Because websites may fail, change structure, or block requests. Proper handling ensures the script continues running.

#### 🔹 16. How do you handle failed HTTP requests?

```python
import requests

try:
    response = requests.get("https://example.com")
    response.raise_for_status()
except requests.exceptions.RequestException as e:
    print("Error:", e)
```

#### 🔹 17. What is logging?

Logging is a way to record events, errors, and program flow for debugging and monitoring.

#### 🔹 18. Basic logging example

```python
import logging

logging.basicConfig(level=logging.INFO)
logging.info("Program started")
logging.error("An error occurred")
```

#### 🔹 19. Difference between print and logging

- `print`: Simple output to console
- `logging`: Structured, configurable, used in real applications

#### 🔹 20. What is debugging?

Debugging is the process of identifying and fixing errors in code.

#### 🔹 21. Common debugging techniques

- Print statements
- Logging
- Using debugger tools (like VS Code debugger)
- Checking stack traces

### Testing (Basic Understanding)


#### 🔹 22. What is unit testing?

Unit testing is testing individual parts (functions/modules) of a program to ensure they work correctly.

#### 🔹 23. Why is testing important?

- Ensures code correctness
- Prevents bugs
- Makes code maintainable
- Helps during refactoring

#### 🔹 24. What is a test case?

A test case is a specific condition or input used to test a function.

#### 🔹 25. Example of a simple unit test (pytest)

```python
def add(a, b):
    return a + b

def test_add():
    assert add(2, 3) == 5
```

#### 🔹 26. What is pytest?

Pytest is a popular Python testing framework used to write simple and scalable test cases.

#### 🔹 27. What is mocking?

Mocking is replacing real objects (like APIs) with fake ones during testing.

#### 🔹 28. Why testing is useful in scraping/automation?

Because scraping scripts break often due to website changes, testing helps quickly detect issues.

#### 🔹 29. What is edge case testing?

Testing unusual or extreme inputs (e.g., empty data, null values).

#### 🔹 30. What happens if you don’t test your code?

- Bugs go unnoticed
- System becomes unreliable
- Hard to maintain and debug
