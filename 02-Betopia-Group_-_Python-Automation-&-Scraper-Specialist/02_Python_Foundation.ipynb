#### 🔹 1. Remove Duplicates from a List

Write a function that removes duplicates from a list without using set(), and keeps the original order.


```python
def remove_duplicates(lst):
    unique_list = []
    
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
    
    return unique_list


# Test
data = [1, 2, 2, 3, 4, 3, 5]
print(remove_duplicates(data))
```

#### 🔹 2. Count Frequency of Words

Given a sentence, count how many times each word appears.


```python
def word_frequency(sentence):
    words = sentence.lower().split()
    freq = {}
    
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    
    return freq


# Test
text = "Python is great and Python is easy"
print(word_frequency(text))
```

#### 🔹 3. List Comprehension Filtering

From a list of numbers, return only even numbers using list comprehension.


```python
def get_even_numbers(nums):
    return [num for num in nums if num % 2 == 0]


# Test
print(get_even_numbers([1, 2, 3, 4, 5, 6]))
```

#### 🔹 4. Exception Handling (Safe Division)

Write a function that divides two numbers and handles division errors properly.


```python
def safe_division(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input type"


# Test
print(safe_division(10, 2))
print(safe_division(10, 0))
```

#### 🔹 5. Read CSV and Process Data

Read a CSV file and return names of people whose age is greater than 25.


```python
import csv

def get_people_above_25(file_path):
    result = []
    
    try:
        with open(file_path, mode='r') as file:
            reader = csv.DictReader(file)
            
            for row in reader:
                if int(row['age']) > 25:
                    result.append(row['name'])
    
    except FileNotFoundError:
        print("File not found")
    
    return result


# Example usage
# print(get_people_above_25("people.csv"))
```

#### 🔹 6. JSON File Handling

Read a JSON file and extract all email addresses.


```python
import json

def extract_emails(file_path):
    emails = []
    
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
            
            for user in data:
                if 'email' in user:
                    emails.append(user['email'])
    
    except Exception as e:
        print("Error:", e)
    
    return emails
```

#### 🔹 8. Dictionary Problem (Top Frequent Word)

Find the most frequent word in a sentence.


```python
def most_frequent_word(sentence):
    words = sentence.lower().split()
    freq = {}
    
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    
    max_word = max(freq, key=freq.get)
    return max_word


# Test
text = "apple banana apple orange banana apple"
print(most_frequent_word(text))
```

#### 🔹 9. Loop + Condition (Filter Data)

From a list of dictionaries, return users with salary > 50,000.


```python
def high_salary_users(users):
    result = []
    
    for user in users:
        if user.get("salary", 0) > 50000:
            result.append(user)
    
    return result


# Test
users = [
    {"name": "A", "salary": 40000},
    {"name": "B", "salary": 60000},
    {"name": "C", "salary": 70000}
]

print(high_salary_users(users))
```

#### 🔹 11. Group Anagrams

Write a function that groups words that are anagrams of each other.

```python
def group_anagrams(words):
    anagrams = {}
    
    for word in words:
        key = ''.join(sorted(word))
        
        if key not in anagrams:
            anagrams[key] = []
        
        anagrams[key].append(word)
    
    return list(anagrams.values())


# Test
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams(words))
```

#### 🔹 12. Flatten Nested List

Write a function to flatten a nested list of arbitrary depth.

```python
def flatten_list(nested):
    result = []
    
    for item in nested:
        if isinstance(item, list):
            result.extend(flatten_list(item))
        else:
            result.append(item)
    
    return result


# Test
data = [1, [2, [3, 4], 5], 6]
print(flatten_list(data))
```

#### 🔹 13. LRU Cache Implementation

Implement a Least Recently Used (LRU) cache using a class.

```python
from collections import OrderedDict

class LRUCache:
    def __init__(self, capacity):
        self.cache = OrderedDict()
        self.capacity = capacity
    
    def get(self, key):
        if key not in self.cache:
            return -1
        
        self.cache.move_to_end(key)
        return self.cache[key]
    
    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)
        
        self.cache[key] = value
        
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)


# Test
cache = LRUCache(2)
cache.put(1, 10)
cache.put(2, 20)
print(cache.get(1))
cache.put(3, 30)
print(cache.get(2))
```

#### 🔹 14. Find First Non-Repeating Character

Write a function to find the first non-repeating character in a string.

```python
def first_non_repeating(s):
    freq = {}
    
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    
    for char in s:
        if freq[char] == 1:
            return char
    
    return None


# Test
print(first_non_repeating("aabbcdeff"))
```

#### 🔹 15. Merge Overlapping Intervals

Given a list of intervals, merge all overlapping intervals.

```python
def merge_intervals(intervals):
    intervals.sort(key=lambda x: x[0])
    merged = []
    
    for interval in intervals:
        if not merged or merged[-1][1] < interval[0]:
            merged.append(interval)
        else:
            merged[-1][1] = max(merged[-1][1], interval[1])
    
    return merged


# Test
intervals = [[1,3],[2,6],[8,10],[15,18]]
print(merge_intervals(intervals))
```

#### 🔹 16. Decorator for Timing Function

Write a decorator that measures execution time of a function.

```python
import time

def time_logger(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        
        result = func(*args, **kwargs)
        
        end = time.time()
        print(f"Execution time: {end - start:.5f} seconds")
        
        return result
    
    return wrapper


@time_logger
def sample_task():
    total = 0
    for i in range(1000000):
        total += i
    return total


# Test
sample_task()
```

#### 🔹 17. Threading (Concurrent Execution)

Write a program to run multiple functions concurrently using threading.

```python
import threading
import time

def task(name):
    print(f"Task {name} started")
    time.sleep(2)
    print(f"Task {name} finished")


threads = []

for i in range(3):
    t = threading.Thread(target=task, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

#### 🔹 18. Find Duplicate Files (Hashing)

Write a function to find duplicate files in a directory using hashing.

```python
import os
import hashlib

def get_file_hash(filepath):
    hasher = hashlib.md5()
    
    with open(filepath, 'rb') as f:
        content = f.read()
        hasher.update(content)
    
    return hasher.hexdigest()


def find_duplicates(folder):
    hashes = {}
    
    for root, _, files in os.walk(folder):
        for file in files:
            path = os.path.join(root, file)
            file_hash = get_file_hash(path)
            
            if file_hash in hashes:
                hashes[file_hash].append(path)
            else:
                hashes[file_hash] = [path]
    
    return [paths for paths in hashes.values() if len(paths) > 1]
```

#### 🔹 20. Retry Mechanism with Exception Handling

Write a function that retries an operation up to 3 times if it fails.

```python
import time

def retry(func, retries=3):
    for attempt in range(retries):
        try:
            return func()
        except Exception as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            time.sleep(1)
    
    return "All attempts failed"


# Example function
def unstable_function():
    import random
    if random.random() < 0.7:
        raise ValueError("Random failure")
    return "Success"


# Test
print(retry(unstable_function))
```
