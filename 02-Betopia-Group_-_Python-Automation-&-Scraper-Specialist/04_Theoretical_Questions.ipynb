### Web Scraping Interview Questions & Answers

**Python Automation and Scraper Specialist**


#### 1. What is web scraping?

Web scraping is the process of automatically extracting data from websites using scripts or tools. It involves sending HTTP requests, retrieving HTML content, and parsing it to extract useful information.


#### 2. What is the difference between web scraping and web crawling?

Web scraping focuses on extracting specific data from web pages, while web crawling involves systematically browsing the web to discover URLs and index content.


#### 3. What is the role of the requests library in Python?

The requests library is used to send HTTP requests (GET, POST, etc.) to a web server and retrieve the response, such as HTML content, JSON data, or status codes.


#### 4. What is BeautifulSoup used for?

BeautifulSoup is a Python library used to parse HTML or XML documents and extract data using tags, classes, IDs, and CSS selectors.


#### 5. What is the difference between .find() and .select() in BeautifulSoup?

.find() is used to get the first matching element using tag and attributes, while .select() uses CSS selectors and can return multiple elements.


#### 6. What are HTML tags, classes, and IDs?

HTML tags define elements (e.g., div, p, a), classes group elements with similar styles or roles, and IDs uniquely identify a specific element on a page.


#### 7. What are CSS selectors?

CSS selectors are patterns used to select HTML elements, such as .class, #id, or tag names. They are commonly used in scraping to locate elements efficiently.


#### 8. What is a User-Agent and why is it important?

A User-Agent is a header sent with HTTP requests that identifies the client (browser or bot). It helps avoid detection and blocking by websites.


#### 9. What is pagination in web scraping?

Pagination refers to navigating through multiple pages of a website to collect data, often by modifying URL parameters or clicking "Next" buttons.


#### 10. What is rate limiting?

Rate limiting is a technique used by websites to restrict the number of requests a client can make in a given time to prevent abuse or overloading.


#### 11. How do you avoid getting blocked while scraping?

You can avoid blocking by using headers, adding delays between requests, rotating IP addresses, using proxies, and mimicking human behavior.


#### 12. What is Selenium and when should you use it?

Selenium is a browser automation tool used to interact with websites. It is used when scraping dynamic websites that rely on JavaScript to load content.


#### 13. What is the difference between static and dynamic websites?

Static websites return fixed HTML content, while dynamic websites load content using JavaScript after the initial page load.


#### 14. What are implicit and explicit waits in Selenium?

Implicit wait sets a global delay for element searches, while explicit wait waits for a specific condition (like element visibility) before proceeding.


#### 15. What is WebDriverWait in Selenium?

WebDriverWait is an explicit wait mechanism that pauses execution until a certain condition is met, such as the presence of an element.


#### 16. What is Scrapy?

Scrapy is a powerful Python framework used for large-scale web scraping. It supports asynchronous requests, making it faster and more efficient.


#### 17. What are the advantages of Scrapy over BeautifulSoup?

Scrapy is faster, handles multiple requests concurrently, has built-in support for pipelines, and is suitable for large-scale scraping projects.


#### 18. What is CAPTCHA and how does it affect scraping?

CAPTCHA is a security measure used to distinguish humans from bots. It prevents automated scraping and may require manual solving or external services.


#### 19. What is robots.txt?

robots.txt is a file on a website that defines rules for web crawlers, specifying which pages can or cannot be accessed.


#### 20. What are common challenges in web scraping?

Common challenges include handling dynamic content, dealing with blocked requests, managing pagination, parsing inconsistent HTML structures, and avoiding detection mechanisms like CAPTCHA.


## 🌐 BeautifulSoup (bs4)


### 🔹 What is BeautifulSoup?

BeautifulSoup is a Python library used for parsing HTML and XML documents. It helps you extract data from static web pages by navigating and searching the HTML structure.

👉 It does NOT fetch data itself — it works with libraries like `requests`.

---

### 🔹 How BeautifulSoup Works

1. Send HTTP request using `requests`
2. Get HTML response
3. Parse HTML using BeautifulSoup
4. Extract data using tags, classes, IDs, or CSS selectors

---

### 🔹 Installation

```bash
pip install beautifulsoup4
pip install requests
```

---

### 🔹 Basic Workflow

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

print(soup.title.text)
```

---

### 🔹 Parsers in BeautifulSoup

- `html.parser` → Built-in (default)
- `lxml` → Faster and recommended
- `html5lib` → More flexible but slower

👉 Interview Tip:
Say: _“I prefer lxml for performance”_

---

### 🔹 Navigating HTML (VERY IMPORTANT)

#### Accessing Elements

```python
soup.title
soup.head
soup.body
```

#### Extract Text

```python
soup.title.text
```

#### Get Attributes

```python
link = soup.find("a")
print(link.get("href"))
```

---

### 🔹 Searching Elements

#### 1. find() → First match

```python
soup.find("div", class_="product")
```

#### 2. find_all() → Multiple matches

```python
soup.find_all("div", class_="product")
```

---

### 🔹 CSS Selectors (VERY IMPORTANT)

```python
soup.select(".class-name")
soup.select("#id-name")
soup.select("div.product")
soup.select("div > a")
```

👉 Interview Tip:
`.select()` is more powerful and flexible

---

### 🔹 Traversing DOM

```python
tag.parent
tag.children
tag.next_sibling
tag.previous_sibling
```

---

### 🔹 Handling Nested Data

```python
products = soup.find_all("div", class_="product")

for p in products:
    name = p.find("h2").text
    price = p.find("span", class_="price").text
```

---

### 🔹 Handling Missing Elements

```python
price = p.find("span", class_="price")

if price:
    print(price.text)
```

---

### 🔹 Cleaning Data

```python
text.strip()
text.replace("\n", "")
```

---

### 🔹 Working with Tables

```python
rows = soup.find_all("tr")

for row in rows:
    cols = row.find_all("td")
```

---

### 🔹 Limitations of BeautifulSoup

❌ Cannot handle JavaScript
❌ Cannot click buttons
❌ Cannot load dynamic content

👉 Only works for **static HTML**

---

### 🔹 When to Use BeautifulSoup

✅ Static websites
✅ Fast scraping needed
✅ Simple data extraction

## 🤖 Selenium


### 🔹 What is Selenium?

Selenium is a browser automation tool used to interact with websites just like a human.

👉 It can:

- Click buttons
- Fill forms
- Scroll pages
- Handle JavaScript

---

### 🔹 Installation

```bash
pip install selenium
```

Also install browser driver (ChromeDriver, etc.)

---

### 🔹 Basic Workflow

```python
from selenium import webdriver

driver = webdriver.Chrome()
driver.get("https://example.com")
```

---

### 🔹 Finding Elements (VERY IMPORTANT)

```python
driver.find_element("id", "id_name")
driver.find_element("class name", "class_name")
driver.find_element("tag name", "div")
```

#### Multiple elements

```python
driver.find_elements("class name", "product")
```

---

### 🔹 Using By (Best Practice)

```python
from selenium.webdriver.common.by import By

driver.find_element(By.ID, "id_name")
driver.find_element(By.CLASS_NAME, "product")
```

---

### 🔹 Interacting with Elements

#### Click

```python
button.click()
```

#### Send input

```python
input_box.send_keys("Hello")
```

---

### 🔹 Waits (VERY IMPORTANT)

#### 1. Implicit Wait

```python
driver.implicitly_wait(10)
```

#### 2. Explicit Wait (Best Practice)

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)

element = wait.until(
    EC.presence_of_element_located((By.CLASS_NAME, "product"))
)
```

---

### 🔹 Handling Dynamic Content

```python
elements = driver.find_elements(By.CLASS_NAME, "product")

for e in elements:
    print(e.text)
```

---

### 🔹 Clicking "Load More"

```python
while True:
    try:
        btn = driver.find_element(By.ID, "load-more")
        btn.click()
    except:
        break
```

---

### 🔹 Scrolling

```python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

---

### 🔹 Handling Alerts

```python
alert = driver.switch_to.alert
alert.accept()
```

---

### 🔹 Handling Multiple Tabs

```python
driver.switch_to.window(driver.window_handles[1])
```

---

### 🔹 Headless Mode (IMPORTANT)

```python
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument("--headless")

driver = webdriver.Chrome(options=options)
```
