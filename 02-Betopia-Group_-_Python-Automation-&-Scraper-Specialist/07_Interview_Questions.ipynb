#### Brief Explanation of My Thesis Work.

My thesis focuses on detecting unusual or suspicious activities in surveillance videos using deep learning. Normally, most systems need examples of abnormal events during training, but in real life, it is very difficult to collect all types of abnormal data. So, in our work, we built a system that can detect anomalies without ever seeing them before. This approach is called zero-shot learning.

The main idea of our model is to first learn what “normal behavior” looks like in a video. Then, whenever something unusual happens, the system can identify it as an anomaly. To do this, we used a combination of three main techniques. First, we used a transformer-based model called TimeSformer to understand both spatial and temporal information from videos, meaning it can understand what is happening and how things change over time. Second, we used a predictive model called DPC-RNN (Deep Predictive Coding with Recurrent Neural Network) that tries to guess future frames of a video. If the prediction is very different from the actual video, it indicates something abnormal. Third, we used a vision-language model called CLIP (Contrastive Language-Image Pre-training) so that the system can understand context using text descriptions, like time, place, or environment.

Another important part of our work is making the model context-aware. This means the system understands that normal behavior can change depending on the situation. For example, running in a park may be normal, but running in a restricted area may be abnormal. My model adjusts its decision based on such context.

Finally, we tested the model on a standard dataset and showed that it performs well compared to other methods, especially in detecting new and unseen anomalies. Overall, my thesis contributes to building a smarter and more practical surveillance system that can work in real-world situations.

### 🟢 Python

#### What is list comprehension?

List comprehension is a concise way to create lists in Python using a single line of code. It improves readability and performance compared to traditional loops.

Example:

```python
squares = [x*x for x in range(5)]
```

#### Difference: list vs tuple

- List: Mutable (can be changed), uses square brackets `[]`
- Tuple: Immutable (cannot be changed), uses parentheses `()`
- Lists are slower but flexible, tuples are faster and safer for fixed data

#### What is a dictionary?

A dictionary is a data structure that stores data in key-value pairs. It allows fast data retrieval using keys.

Example:

```python
data = {"name": "Tareq", "age": 22}
```

### 🟢 Scraping

#### What is BeautifulSoup?

BeautifulSoup is a Python library used for parsing HTML and XML documents. It helps extract data from web pages by navigating the HTML structure.

#### Difference: Selenium vs BeautifulSoup

- BeautifulSoup:
  - Parses static HTML
  - Faster
  - Does not execute JavaScript

- Selenium:
  - Automates browsers
  - Handles dynamic content (JavaScript)
  - Slower but more powerful

#### What is User-Agent?

User-Agent is a string sent in HTTP headers that identifies the browser and device making the request. Websites use it to detect bots or allow/block requests.

#### Why scraping fails?

- Website structure changes
- JavaScript-loaded content
- IP blocking or rate limiting
- Missing headers (like User-Agent)
- CAPTCHA or anti-bot protection

### 🟢 API

#### What is REST API?

A REST API is a way for applications to communicate over HTTP using standard methods like GET and POST. It returns data usually in JSON format.

#### Difference between GET and POST

- GET:
  - Used to retrieve data
  - Parameters are sent in URL
  - No side effects

- POST:
  - Used to send data to server
  - Data is sent in request body
  - Can modify server data
