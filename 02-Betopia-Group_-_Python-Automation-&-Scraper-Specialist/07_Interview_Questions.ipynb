### 🟢 Python

#### What is list comprehension?

List comprehension is a concise way to create lists in Python using a single line of code. It improves readability and performance compared to traditional loops.

Example:

```python
squares = [x*x for x in range(5)]
```

#### Difference: list vs tuple

- List: Mutable (can be changed), uses square brackets `[]`
- Tuple: Immutable (cannot be changed), uses parentheses `()`
- Lists are slower but flexible, tuples are faster and safer for fixed data

#### What is a dictionary?

A dictionary is a data structure that stores data in key-value pairs. It allows fast data retrieval using keys.

Example:

```python
data = {"name": "Tareq", "age": 22}
```

### 🟢 Scraping

#### What is BeautifulSoup?

BeautifulSoup is a Python library used for parsing HTML and XML documents. It helps extract data from web pages by navigating the HTML structure.

#### Difference: Selenium vs BeautifulSoup

- BeautifulSoup:
  - Parses static HTML
  - Faster
  - Does not execute JavaScript

- Selenium:
  - Automates browsers
  - Handles dynamic content (JavaScript)
  - Slower but more powerful

#### What is User-Agent?

User-Agent is a string sent in HTTP headers that identifies the browser and device making the request. Websites use it to detect bots or allow/block requests.

#### Why scraping fails?

- Website structure changes
- JavaScript-loaded content
- IP blocking or rate limiting
- Missing headers (like User-Agent)
- CAPTCHA or anti-bot protection

### 🟢 API

#### What is REST API?

A REST API is a way for applications to communicate over HTTP using standard methods like GET and POST. It returns data usually in JSON format.

#### Difference between GET and POST

- GET:
  - Used to retrieve data
  - Parameters are sent in URL
  - No side effects

- POST:
  - Used to send data to server
  - Data is sent in request body
  - Can modify server data
