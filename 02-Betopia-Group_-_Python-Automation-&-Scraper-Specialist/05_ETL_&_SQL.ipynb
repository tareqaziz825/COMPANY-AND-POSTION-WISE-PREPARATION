### Data Processing (ETL)

#### 🔹 1. What is ETL and why is it important?
ETL stands for Extract, Transform, Load. It is used to collect data from sources, clean/transform it into a usable format, and store it in a database or file system. It is important because raw data is often messy and unusable without processing.

#### 🔹 2. What are common data cleaning steps in Python?
Common steps include:
- Removing duplicates
- Handling missing values
- Fixing inconsistent formats
- Removing unwanted characters
- Converting data types


#### 🔹 3. How do you handle missing values in a dataset?
